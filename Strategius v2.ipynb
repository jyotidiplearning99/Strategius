````python
import os
import json
import uuid
import logging
from datetime import datetime
from IPython.display import display, HTML
import ipywidgets as widgets
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
import pandas as pd
import re
import json

# --- Logging ---
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)


# --- API KEY Setup ---
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "")
OPENAI_API_KEY="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"
if not OPENAI_API_KEY:
    api_key_widget = widgets.Password(description='üîë OpenAI API Key:', placeholder='Paste API Key', layout=widgets.Layout(width='50%'))

    def save_api_key(change):
        os.environ["OPENAI_API_KEY"] = change['new']
        global OPENAI_API_KEY
        OPENAI_API_KEY = change['new']
        print("‚úÖ API key saved!")

    api_key_widget.observe(save_api_key, names='value')
    display(api_key_widget)

# --- Save to Excel ---
def save_to_excel(experience_level, answers, strategy, filename='strategy_user_data.xlsx'):
    user_id = str(uuid.uuid4())
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    row = {
        'User ID': user_id,
        'Timestamp': timestamp,
        'Experience Level': experience_level,
        'Answers': "; ".join([f"{a['angle']}: {a['answer']}" for a in answers]),
        'Strategy Title': strategy.get('title', 'N/A'),
        'Description': strategy.get('description', 'N/A'),
        'Steps': "; ".join(strategy.get('steps', [])),
        'Outcomes': "; ".join(strategy.get('outcomes', [])),
        'Metrics': "; ".join(strategy.get('metrics', [])),
        'Action Items': "; ".join(strategy.get('action_items', [])),
        'Confidence': strategy.get('confidence', 'N/A')
    }

    try:
        existing_df = pd.read_excel(filename)
        df = pd.concat([existing_df, pd.DataFrame([row])], ignore_index=True)
    except FileNotFoundError:
        df = pd.DataFrame([row])

    df.to_excel(filename, index=False)
    print("‚úÖ Strategy data saved to", filename)

# --- Example Questions ---
QUESTIONS = [
    {"id": 1, "angle": "Competitive Advantage", "question": "What is your key competitive advantage?", "importance": "Clarifies your Unique Selling Proposition (USP)."},
    {"id": 2, "angle": "Target Market", "question": "Who are your ideal customers?", "importance": "Defines focus areas to increase customer alignment."},
    {"id": 3, "angle": "Business Model", "question": "How do you generate revenue?", "importance": "Ensures sustainability and profitability."},
]

# --- Workflow Logic ---
class StrategyAssessmentWorkflow:
    def get_initial_question(self):
        return self.experience_level_question()

    def experience_level_question(self):
        return {
            "question": "How familiar are you with strategy work?",
            "type": "scale",
            "options": [
                {"value": 1, "label": "New to strategy"},
                {"value": 2, "label": "Some experience"},
                {"value": 3, "label": "Experienced"},
                {"value": 4, "label": "Expert"}
            ]
        }

    def process_experience_level(self, level):
        return {
            "question": "Does your company have a formal business strategy?",
            "type": "yes_no",
            "experience_level": level
        }

    def process_has_strategy(self, has_strategy, experience_level):
        msg = "Answer these strategy refinement questions." if has_strategy else "Let's help you build your strategy!"
        return {"message": msg, "questions": QUESTIONS}

# --- Main UI ---
class StrategyAssessmentUI:
    def __init__(self):
        self.workflow = StrategyAssessmentWorkflow()
        self.state = {"step": "initial", "data": {}}
        self.output = widgets.Output()

    def start(self):
        with self.output:
            self.output.clear_output()
            question = self.workflow.get_initial_question()
            self._display_question(question)
        return self.output

    def _display_question(self, question_data):
        display(widgets.HTML(f"<h3 style='color:#2c3e50;'>{question_data['question']}</h3>"))

        if question_data['type'] == 'scale':
            self._display_scale_buttons(question_data['options'])
        elif question_data['type'] == 'yes_no':
            self._display_yes_no_buttons(question_data)

    def _display_scale_buttons(self, options):
        buttons = []
        for opt in options:
            btn = widgets.Button(description=f"{opt['value']}: {opt['label']}", layout=widgets.Layout(width='300px'))
            btn.on_click(self._create_scale_callback(opt['value']))
            buttons.append(btn)
        display(widgets.VBox(buttons))

    def _create_scale_callback(self, val):
        def callback(b):
            self._handle_experience_level(val)
        return callback

    def _display_yes_no_buttons(self, question_data):
        yes_btn = widgets.Button(description='Yes', button_style='success', layout=widgets.Layout(width='100px'))
        no_btn = widgets.Button(description='No', button_style='danger', layout=widgets.Layout(width='100px'))
        info_btn = widgets.Button(description='‚ìò', layout=widgets.Layout(width='30px', height='30px'))

        yes_btn.on_click(lambda b: self._handle_yes_no(True, question_data))
        no_btn.on_click(lambda b: self._handle_yes_no(False, question_data))
        info_btn.on_click(lambda b: self._show_hint("Having a formal strategy guides your business growth."))

        display(widgets.HBox([yes_btn, no_btn, info_btn]))

    def _show_hint(self, message):
        display(widgets.HTML(f"<div style='color:#888;'>{message}</div>"))

    def _handle_experience_level(self, level):
        self.state['data']['experience_level'] = level
        self.state['step'] = 'experience_level'
        question = self.workflow.process_experience_level(level)
        with self.output:
            self.output.clear_output()
            self._display_question(question)

    def _handle_yes_no(self, answer, question_data):
        self.state['data']['has_strategy'] = answer
        self.state['step'] = 'has_strategy'
        data = self.workflow.process_has_strategy(answer, self.state['data']['experience_level'])
        with self.output:
            self.output.clear_output()
            self._display_questions_form(data)

    def _display_questions_form(self, data):
        display(widgets.HTML(f"<h3 style='color:#34495e;'>{data['message']}</h3>"))

        form_items = []
        answer_widgets = []

        for idx, q in enumerate(data['questions']):
            question_html = widgets.HTML(f"<b>{idx + 1}. {q['angle']}</b><br>{q['question']}")
            hint_btn = widgets.Button(description='‚ìò', layout=widgets.Layout(width='30px', height='30px'))

            hint_btn.on_click(self._make_hint_callback(q['importance']))

            answer_box = widgets.Textarea(placeholder='Type your answer...', layout=widgets.Layout(width='100%', height='60px'))

            form_items.append(widgets.HBox([question_html, hint_btn]))
            form_items.append(answer_box)
            answer_widgets.append(answer_box)

        submit_btn = widgets.Button(description='Generate Strategy', button_style='primary', layout=widgets.Layout(width='300px'))

        submit_btn.on_click(lambda b: self._on_submit_answers(answer_widgets, data['questions']))

        display(widgets.VBox(form_items + [submit_btn]))

    def _make_hint_callback(self, hint_text):
        def callback(b):
            self._show_hint(hint_text)
        return callback

    def _on_submit_answers(self, answer_widgets, questions):
        answers = []
        for idx, q in enumerate(questions):
            answers.append({
                'question': q['question'],
                'answer': answer_widgets[idx].value.strip() or 'N/A',
                'angle': q['angle'],
                'importance': q['importance']
            })

        self.state['data']['answers'] = answers
        self._generate_strategy()

    def _generate_strategy(self):
        with self.output:
            self.output.clear_output()
            display(widgets.HTML("<h4 style='color:#2980b9;'>Generating strategy... please wait!</h4>"))

        engine = StrategyEngine()
        qa_pairs = [{'question': ans['question'], 'answer': ans['answer']} for ans in self.state['data']['answers']]
        experience_level = self.state['data'].get('experience_level', 'Unknown')
        constraints = [
            'Has strategy' if self.state['data'].get('has_strategy') else 'No strategy',
            f"Experience level: {experience_level}"
        ]

        try:
            strategy = engine.generate_strategy(qa_pairs, constraints)
            save_to_excel(experience_level, self.state['data']['answers'], strategy)

            with self.output:
                self.output.clear_output()
                display(display_strategy_html(strategy))

        except Exception as e:
            with self.output:
                self.output.clear_output()
                display(widgets.HTML(f"<div style='color:red;'>‚ùå Error generating strategy: {str(e)}</div>"))

# --- Engine ---
class StrategyEngine:
    def __init__(self):
        if not OPENAI_API_KEY:
            raise ValueError("OpenAI API key not set!")
        self.llm = ChatOpenAI(api_key=OPENAI_API_KEY, temperature=0.5, model="gpt-4")
        self.prompt = ChatPromptTemplate.from_template("""
You are an expert business strategist. 
Based on the user's constraints and QA, provide ONLY a valid JSON object with the following fields:
- title
- description
- steps (list of step descriptions)
- outcomes (list of outcomes)
- metrics (list of measurable KPIs)
- action_items (list of immediate actions)
- confidence (confidence score between 1 and 10)

DO NOT include any explanations, comments, or markdown formatting. Output strictly a JSON object.
Constraints:
{constraints}

QA Pairs:
{qa_pairs}
""")
        self.chain = self.prompt | self.llm

    def clean_json(self, response_text):
        """
        Cleans up the raw LLM output to extract a valid JSON string.
        Removes markdown code blocks and extra text.
        """
        # Remove markdown code block if present
        cleaned = re.sub(r"```(json)?", "", response_text).strip()
        # Find the first valid JSON-like structure
        match = re.search(r"\{.*\}", cleaned, re.DOTALL)
        if match:
            return match.group(0)
        else:
            raise ValueError("No valid JSON found in LLM response.")

    def generate_strategy(self, qa_pairs, constraints):
        formatted_qa = "\n".join([f"Q: {qa['question']}\nA: {qa['answer']}" for qa in qa_pairs])
        formatted_constraints = "\n".join([f"- {c}" for c in constraints])

        logger.info("Sending prompt to GPT...")
        result = self.chain.invoke({
            "qa_pairs": formatted_qa,
            "constraints": formatted_constraints
        })

        response_text = result.content
        logger.info("Raw response from GPT: %s", response_text)

        # Clean the response before loading JSON
        try:
            cleaned_json = self.clean_json(response_text)
            logger.info("Clean JSON: %s", cleaned_json)
            return json.loads(cleaned_json)
        except Exception as e:
            logger.error(f"Failed to parse strategy JSON: {str(e)}")
            raise e

# --- Display Strategy ---
def display_strategy_html(strategy):
    title = strategy.get('title', 'Strategy Plan')
    description = strategy.get('description', 'No description provided')
    steps = strategy.get('steps', ["No steps provided"])
    outcomes = strategy.get('outcomes', ["No outcomes provided"])
    metrics = strategy.get('metrics', ["No metrics provided"])
    action_items = strategy.get('action_items', ["No action items provided"])
    confidence = strategy.get('confidence', 'N/A')

    def ul_list(items):
        return ''.join(f"<li>{item}</li>" for item in items)

    html = f"""
    <div style='padding:20px; border:1px solid #ccc; border-radius:10px;'>
        <h2 style='color:#2c3e50;'>{title}</h2>
        <p><strong>Description:</strong> {description}</p>

        <h3>Steps</h3><ul>{ul_list(steps)}</ul>
        <h3>Outcomes</h3><ul>{ul_list(outcomes)}</ul>
        <h3>Metrics</h3><ul>{ul_list(metrics)}</ul>
        <h3>Action Items</h3><ul>{ul_list(action_items)}</ul>
        <p><strong>Confidence:</strong> {confidence}</p>
    </div>
    """
    return HTML(html)

# --- Start UI ---
ui = StrategyAssessmentUI()
display(ui.start())
````

````python
import os
import json
import uuid
import logging
from datetime import datetime
from IPython.display import display, HTML
import ipywidgets as widgets
import pandas as pd
import re
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

# --- Logging ---
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# --- API KEY Setup ---
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "")
OPENAI_API_KEY="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"
if not OPENAI_API_KEY:
    api_key_widget = widgets.Password(
        description='üîë OpenAI API Key:',
        placeholder='Paste API Key',
        layout=widgets.Layout(width='50%')
    )

    def save_api_key(change):
        os.environ["OPENAI_API_KEY"] = change['new']
        
        global OPENAI_API_KEY
        OPENAI_API_KEY = change['new']
        print("‚úÖ API key saved!")

    api_key_widget.observe(save_api_key, names='value')
    display(api_key_widget)

# --- YES / NO QUESTIONS ---
YES_QUESTIONS = [
    {
        "id": 1,
        "angle": "Competitive advantage",
        "question": "What is your key competitive advantage, and how do you differentiate yourself from competitors?",
        "importance": "Clarifies your Unique Selling Proposition (USP) and sustainability."
    },
    {
        "id": 2,
        "angle": "Target market",
        "question": "Who exactly are your target customers, and how does your product meet their needs?",
        "importance": "Defines focus areas to increase customer alignment."
    },
    {
        "id": 3,
        "angle": "Business model",
        "question": "What is your revenue model and why is it best for your market?",
        "importance": "Ensures sustainability and profitability."
    },
    {
        "id": 4,
        "angle": "Market size and growth",
        "question": "What is your company's scalability and growth potential in the next 3‚Äì5 years?",
        "importance": "Evaluates scalability and feasibility of growth."
    },
    {
        "id": 5,
        "angle": "Profitability",
        "question": "What are the short-term and long-term prospects for profitability?",
        "importance": "Assesses financial sustainability and viability."
    },
    {
        "id": 6,
        "angle": "Objectives and metrics",
        "question": "How do you measure success, and what are your KPIs?",
        "importance": "Clarifies measurement and monitoring effectiveness."
    },
    {
        "id": 7,
        "angle": "Risks",
        "question": "What are the most significant risks in your business and how are you mitigating them?",
        "importance": "Identifies major risks and risk management."
    },
    {
        "id": 8,
        "angle": "Market change",
        "question": "What trends or changes in your operating environment could impact your business?",
        "importance": "Assesses adaptability to external changes."
    },
    {
        "id": 9,
        "angle": "Customer strategy",
        "question": "What is your customer acquisition and retention strategy?",
        "importance": "Reveals customer management and loyalty focus."
    },
    {
        "id": 10,
        "angle": "Leadership",
        "question": "What is your leadership model, and how do you ensure team alignment?",
        "importance": "Links leadership to execution of company goals."
    }
]

NO_QUESTIONS = [
    {
        "id": 1,
        "angle": "Purpose and vision",
        "question": "Why does your company exist and what is its long-term vision?",
        "importance": "Clarifies company purpose and long-term goals."
    },
    {
        "id": 2,
        "angle": "Target market",
        "question": "Who are your customers, and what problems are you solving for them?",
        "importance": "Identifies target audience and pain points."
    },
    {
        "id": 3,
        "angle": "Competitive advantage",
        "question": "What makes your company unique in the market?",
        "importance": "Defines differentiation from competitors."
    },
    {
        "id": 4,
        "angle": "Products/services",
        "question": "What products or services do you offer, and what value do they deliver?",
        "importance": "Clarifies offering and customer value."
    },
    {
        "id": 5,
        "angle": "Business model",
        "question": "How does your company make money?",
        "importance": "Explains the revenue and financial model."
    },
    {
        "id": 6,
        "angle": "Market size and growth",
        "question": "How large is your target market and its growth potential?",
        "importance": "Determines market opportunity."
    },
    {
        "id": 7,
        "angle": "Key assets",
        "question": "What key resources or partnerships are essential for success?",
        "importance": "Outlines critical capabilities and alliances."
    },
    {
        "id": 8,
        "angle": "Objectives and metrics",
        "question": "What measurable objectives do you have for the next 1‚Äì3 years?",
        "importance": "Clarifies goals and success tracking."
    },
    {
        "id": 9,
        "angle": "Risks",
        "question": "What are your key risks and how are you managing them?",
        "importance": "Explores risk management strategies."
    },
    {
        "id": 10,
        "angle": "Action steps",
        "question": "What are your next key action steps and responsibilities?",
        "importance": "Defines immediate priorities and accountability."
    }
]

# --- Save to Excel ---
def save_to_excel(experience_level, answers, strategy, filename='strategy_user_data.xlsx'):
    user_id = str(uuid.uuid4())
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    row = {
        'User ID': user_id,
        'Timestamp': timestamp,
        'Experience Level': experience_level,
        'Answers': "; ".join([f"{a['angle']}: {a['answer']}" for a in answers]),
        'Strategy Title': strategy.get('title', 'N/A'),
        'Description': strategy.get('description', 'N/A'),
        'Steps': "; ".join(strategy.get('steps', [])),
        'Outcomes': "; ".join(strategy.get('outcomes', [])),
        'Metrics': "; ".join(strategy.get('metrics', [])),
        'Action Items': "; ".join(strategy.get('action_items', [])),
        'Confidence': strategy.get('confidence', 'N/A')
    }

    try:
        existing_df = pd.read_excel(filename)
        df = pd.concat([existing_df, pd.DataFrame([row])], ignore_index=True)
    except FileNotFoundError:
        df = pd.DataFrame([row])

    df.to_excel(filename, index=False)
    print("‚úÖ Strategy data saved to", filename)

# --- Workflow Logic ---
class StrategyAssessmentWorkflow:
    def get_initial_question(self):
        return self.experience_level_question()

    def experience_level_question(self):
        return {
            "question": "How familiar are you with strategy work?",
            "type": "scale",
            "options": [
                {"value": 1, "label": "New to strategy"},
                {"value": 2, "label": "Some experience"},
                {"value": 3, "label": "Experienced"},
                {"value": 4, "label": "Expert"}
            ]
        }

    def process_experience_level(self, level):
        return {
            "question": "Does your company have a formal business strategy?",
            "type": "yes_no",
            "experience_level": level
        }

    def process_has_strategy(self, has_strategy, experience_level):
        msg = "Answer these strategy refinement questions." if has_strategy else "Let's help you build your strategy!"
        questions = YES_QUESTIONS if has_strategy else NO_QUESTIONS
        return {"message": msg, "questions": questions}

# --- Main UI ---
class StrategyAssessmentUI:
    def __init__(self):
        self.workflow = StrategyAssessmentWorkflow()
        self.state = {"step": "initial", "data": {}}
        self.output = widgets.Output()

    def start(self):
        with self.output:
            self.output.clear_output()
            question = self.workflow.get_initial_question()
            self._display_question(question)
        return self.output

    def _display_question(self, question_data):
        display(widgets.HTML(f"<h3 style='color:#2c3e50;'>{question_data['question']}</h3>"))

        if question_data['type'] == 'scale':
            self._display_scale_buttons(question_data['options'])
        elif question_data['type'] == 'yes_no':
            self._display_yes_no_buttons(question_data)

    def _display_scale_buttons(self, options):
        buttons = []
        for opt in options:
            btn = widgets.Button(description=f"{opt['value']}: {opt['label']}", layout=widgets.Layout(width='300px'))
            btn.on_click(self._create_scale_callback(opt['value']))
            buttons.append(btn)
        display(widgets.VBox(buttons))

    def _create_scale_callback(self, val):
        def callback(b):
            self._handle_experience_level(val)
        return callback

    def _display_yes_no_buttons(self, question_data):
        yes_btn = widgets.Button(description='Yes', button_style='success', layout=widgets.Layout(width='100px'))
        no_btn = widgets.Button(description='No', button_style='danger', layout=widgets.Layout(width='100px'))
        info_btn = widgets.Button(description='‚ìò', layout=widgets.Layout(width='30px', height='30px'))

        yes_btn.on_click(lambda b: self._handle_yes_no(True, question_data))
        no_btn.on_click(lambda b: self._handle_yes_no(False, question_data))
        info_btn.on_click(lambda b: self._show_hint("Formal strategies help guide company growth."))

        display(widgets.HBox([yes_btn, no_btn, info_btn]))

    def _show_hint(self, message):
        with self.output:
            display(widgets.HTML(f"<p style='color:#888;'>{message}</p>"))

    def _handle_experience_level(self, level):
        self.state['data']['experience_level'] = level
        self.state['step'] = 'experience_level'
        question = self.workflow.process_experience_level(level)
        with self.output:
            self.output.clear_output()
            self._display_question(question)

    def _handle_yes_no(self, answer, question_data):
        self.state['data']['has_strategy'] = answer
        self.state['step'] = 'has_strategy'
        data = self.workflow.process_has_strategy(answer, self.state['data']['experience_level'])
        with self.output:
            self.output.clear_output()
            self._display_questions_form(data)

    def _display_questions_form(self, data):
        display(widgets.HTML(f"<h3 style='color:#34495e;'>{data['message']}</h3>"))

        form_items = []
        answer_widgets = []

        for idx, q in enumerate(data['questions']):
            question_html = widgets.HTML(f"<b>{idx + 1}. {q['angle']}</b><br>{q['question']}")
            hint_output = widgets.Output()

            hint_btn = widgets.Button(description='‚ìò', layout=widgets.Layout(width='30px', height='30px'))
            hint_btn.on_click(self._make_hint_callback(q['importance'], hint_output))

            answer_box = widgets.Textarea(placeholder='Type your answer...', layout=widgets.Layout(width='100%', height='80px'))

            form_items.append(widgets.VBox([
                widgets.HBox([question_html, hint_btn]),
                hint_output,
                answer_box
            ]))
            answer_widgets.append(answer_box)

        submit_btn = widgets.Button(description='Generate Strategy', button_style='primary', layout=widgets.Layout(width='300px'))
        submit_btn.on_click(lambda b: self._on_submit_answers(answer_widgets, data['questions']))

        display(widgets.VBox(form_items + [submit_btn]))

    def _make_hint_callback(self, hint_text, hint_output_widget):
        def callback(b):
            with hint_output_widget:
                hint_output_widget.clear_output()
                display(widgets.HTML(f"<div style='color:gray; font-style:italic;'>{hint_text}</div>"))
        return callback

    def _on_submit_answers(self, answer_widgets, questions):
        answers = []
        for idx, q in enumerate(questions):
            answers.append({
                'question': q['question'],
                'answer': answer_widgets[idx].value.strip() or 'N/A',
                'angle': q['angle'],
                'importance': q['importance']
            })

        self.state['data']['answers'] = answers
        self._generate_strategy()

    def _generate_strategy(self):
        with self.output:
            self.output.clear_output()
            display(widgets.HTML("<h4 style='color:#2980b9;'>Generating strategy... please wait!</h4>"))

        engine = StrategyEngine()
        qa_pairs = [{'question': ans['question'], 'answer': ans['answer']} for ans in self.state['data']['answers']]
        experience_level = self.state['data'].get('experience_level', 'Unknown')
        constraints = [
            'Has strategy' if self.state['data'].get('has_strategy') else 'No strategy',
            f"Experience level: {experience_level}"
        ]

        try:
            strategy = engine.generate_strategy(qa_pairs, constraints)
            save_to_excel(experience_level, self.state['data']['answers'], strategy)

            with self.output:
                self.output.clear_output()
                display(display_strategy_html(strategy))

        except Exception as e:
            with self.output:
                self.output.clear_output()
                display(widgets.HTML(f"<div style='color:red;'>‚ùå Error generating strategy: {str(e)}</div>"))

# --- Strategy Engine ---
class StrategyEngine:
    def __init__(self):
        if not OPENAI_API_KEY:
            raise ValueError("OpenAI API key not set!")
        self.llm = ChatOpenAI(api_key=OPENAI_API_KEY, temperature=0.5, model="gpt-4")
        self.prompt = ChatPromptTemplate.from_template("""
You are an expert business strategist.
Based on the user's constraints and QA, provide ONLY a valid JSON object with the following fields:
- title
- description
- steps
- outcomes
- metrics
- action_items
- confidence

DO NOT include any explanations, comments, or markdown formatting. Output strictly JSON.
Constraints:
{constraints}

QA Pairs:
{qa_pairs}
""")
        self.chain = self.prompt | self.llm

    def generate_strategy(self, qa_pairs, constraints):
        formatted_qa = "\n".join([f"Q: {qa['question']}\nA: {qa['answer']}" for qa in qa_pairs])
        formatted_constraints = "\n".join([f"- {c}" for c in constraints])

        logger.info("Sending prompt to GPT...")
        result = self.chain.invoke({"qa_pairs": formatted_qa, "constraints": formatted_constraints})

        cleaned_json = re.sub(r"```(?:json)?", "", result.content).strip()
        return json.loads(cleaned_json)

# --- Display Strategy ---
def display_strategy_html(strategy):
    def ul_list(items):
        return ''.join(f"<li>{item}</li>" for item in items)

    html = f"""
    <div style='padding:20px; border:1px solid #ccc; border-radius:10px;'>
        <h2 style='color:#2c3e50;'>{strategy.get('title', 'Strategy Plan')}</h2>
        <p><strong>Description:</strong> {strategy.get('description', 'No description provided')}</p>

        <h3>Steps</h3><ul>{ul_list(strategy.get('steps', []))}</ul>
        <h3>Outcomes</h3><ul>{ul_list(strategy.get('outcomes', []))}</ul>
        <h3>Metrics</h3><ul>{ul_list(strategy.get('metrics', []))}</ul>
        <h3>Action Items</h3><ul>{ul_list(strategy.get('action_items', []))}</ul>
        <p><strong>Confidence:</strong> {strategy.get('confidence', 'N/A')}</p>
    </div>
    """
    return HTML(html)

# --- Start UI ---
ui = StrategyAssessmentUI()
display(ui.start())
````
